```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
df.head()
```

using map to change the gender column

```python
df['Gender']=df['Gender'].map({'Male':1,'Female':0})
```

```python
df['Gender']
```

```python
from sklearn.model_selection import train_test_split 
x=df.drop('Purchased',axis=1)
y=df['Purchased']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train_scalled=ss.fit_transform(x_train)
x_test_scalled=ss.transform(x_test)
```

```python
x_train_scalled
```

```python
x_test_scalled
```

```python
x_train_scalled=pd.DataFrame(data=x_train_scalled,columns=x_train.columns)
x_test_scalled=pd.DataFrame(data=x_test_scalled,columns=x_train.columns)
```

```python
x_train_scalled
```

```python
x_test_scalled
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(x_train['Age'],x_train['EstimatedSalary'])
plt.title('Before Scalling')
plt.xlabel('Age')
plt.ylabel('Salary')

plt.subplot(1,2,2)
plt.scatter(x_train_scalled['Age'],x_train_scalled['EstimatedSalary'],color='Red')
plt.title('After Scalling')
plt.xlabel('Age')
plt.ylabel('Salary')

plt.tight_layout()
plt.show()
```
