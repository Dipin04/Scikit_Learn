importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

loading datasets

```python
df=pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv',usecols=['name','fuel','seller_type','transmission','owner'])
df.head()
```

```python
df.drop('name',axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x=df.drop('fuel',axis=1)
y=df['fuel']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
oe=OrdinalEncoder()
le=LabelEncoder()
```

```python
y_train_scalled=le.fit_transform(y_train)
y_test_scalled=le.transform(y_test)
```

```python
x_train_scalled=oe.fit_transform(x_train)
x_test_scalled=oe.transform(x_test)
```

```python
x_train_scalled
```

```python
x_train_scalled=pd.DataFrame(data=x_train_scalled,columns=x_train.columns)
x_test_scalled=pd.DataFrame(data=x_test_scalled,columns=x_train.columns)
```
