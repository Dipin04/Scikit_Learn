importing libraries

```python
import pandas as pd
import numpy as np
```

Loading Datasets

```python
df=pd.read_excel('onedata.xlsx')
df.head()
                 
```

Using python get dummies function to do one hot encoder

```python
pd.get_dummies(data=df,columns=['buyer','fruits','gender'],dtype=int)
```

using scikit learn to use onehotencoder

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',dtype=int,sparse_output=False)
```

```python
df_transformed=ohe.fit_transform(df[['buyer','fruits','gender']])
```

```python
df_transformed
```

```python
cols=ohe.get_feature_names_out()
pd.DataFrame(data=df_transformed,columns=cols)
```
